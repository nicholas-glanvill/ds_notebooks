#### Exercise 1:
Write a function that takes a whole number and returns the sum of its digits. Name it `sum_of_digits`.

````python
# Your code here
def sum_of_digits(num):
    str_num = str(num)
    str_sum = 0
    for digit in str_num:
        str_sum += int(digit)
    return str_sum
````

````python
# Test your function
sum_of_digits(12345)

# Expected output: 15
````

#### Exercise 2:
Given a dictionary     
```
    {'place': 'window',
     'who': 'Juliet'}
```    
write a function that returns the string   
```
'But, soft! what light through yonder window breaks?
It is the east, and Juliet is the sun.'
```     
Name it `light_breaks`.

````python
# Your code here
def light_breaks(my_dict):
    return f'''But, soft! what light through yonder {my_dict['place']} breaks?
It is the east, and {my_dict['who']} is the sun.'''
````

````python
# Test your function
test_dict = {'place': 'monitor',
             'who': 'StackOverflow'}
print(light_breaks(test_dict))
````

#### Exercise 3
Modify your function so it will fill in 'window', or 'Juliet' in the appropriate place if the argument given does not have the proper keys.    
E.g.
```
light_breaks_2(
    {'occupation': 'rich duck',
     'who':'Scrooge McDuck',
     'accent': 'Scottish'})
```
should have the output
```
'But, soft! what light through yonder window breaks?
It is the east, and Scrooge McDuck is the sun.'
```

````python
def light_breaks_2(my_dict):
    default_dict ={'place': 'window',
                   'who': 'Juliet'}
    default_dict.update(my_dict)
    return f'''But, soft! what light through yonder {default_dict['place']} breaks?
It is the east, and {default_dict['who']} is the sun.'''
````

````python
# Test your function
print(light_breaks_2(
    {'occupation': 'rich duck',
     'who':'Scrooge McDuck',
     'accent': 'Scottish'}))
````

#### Exercise 4

LEGO sets from the late 1990s have been determined to be valuable to collectors. Likewise, Advent Calendar sets have become popular among these same collectors. Naturally, sets fulfilling both of these conditions are worth even more.

Write a function that takes a Series (one row of `sets_df`) and returns     
- 'valuable' if the Series fulfills one of the conditions above
- 'very valuable' if it fulfills both conditions
- 'not valuable' if it fulfills neither

Name the function `is_valuable`.

````python
# Run this cell to get the data
import pandas as pd
url = 'https://drive.google.com/file/d/1P2EyVCFAPeJcQ_lAN2W-bjwtbizOeFXE/view?usp=drive_link'
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
sets_df = pd.read_csv(path)
````

````python
# Your code here
def is_valuable(item):
    is_90s = item['year'] >= 1995 and item['year'] < 2000
    is_advent = item['name'].startswith('Advent Calendar')
    if is_90s and is_advent:
        return 'very valuable'
    elif is_90s or is_advent:
        return 'valuable'
    else:
        return 'not valuable'
````

````python
# Test your function
print(is_valuable(sets_df.loc[0]))     # 'not valuable'
print(is_valuable(sets_df.loc[1]))     # 'valuable'
print(is_valuable(sets_df.loc[499]))     # 'very valuable'
````
